```python
# Crear una nueva columna llamada 'salary_group' que clasifique a cada empleado según su salario
df['salary_group'] = pd.cut(
    df['salary_in_usd'],                           # Columna de salarios anuales en USD
    bins=[0, 100000, 250000, float('inf')],        # Rangos definidos: 0-100K, 100K-250K, más de 250K
    labels=['0-100K USD', '100K-250K USD', '250K+ USD'],  # Etiquetas para los rangos
    include_lowest=True                            # Incluir el valor mínimo (0) en el primer grupo
)

# Definir el orden lógico de los grupos salariales (de menor a mayor)
group_order = ['0-100K USD', '100K-250K USD', '250K+ USD']

# Contar cuántos empleados hay en cada grupo salarial
salary_counts = df['salary_group'].value_counts()     # Cuenta la frecuencia de cada grupo
salary_counts = salary_counts.reindex(group_order)    # Reordena los grupos según el orden definido


plt.figure(figsize=(8, 4))                             # Crear una nueva figura con tamaño 8x4 pulgadas

salary_counts.plot(kind='bar', color='purple')          # Crear gráfico de barras con color verde

plt.title('Número de empleados por rango salarial')     # Título del gráfico
plt.ylabel('Cantidad de empleados')                    # Etiqueta del eje Y
plt.xlabel('Rango salarial')                           # Etiqueta del eje X
plt.grid(axis='y')                                     # Mostrar cuadrícula horizontal
plt.show()                                             # Mostrar el gráfico en pantalla
```

## 5. Estadistica Avanzada

Calcular la medidas de tendencia central, de nuestro dataFrame

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, mode

def calculate_statistics(df):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    stat_names = [
        'Mean', 'Median', 'Mode', 'Range', 'Stdev', 'Variance', 'IQR',
        '25%', '50%', '75%', 'Min', 'Max', 'Skewness', 'Kurtosis'
    ]

    stats_table = []

    for stat in stat_names:
        row = []
        for col in numeric_cols:
            data = df[col].dropna()
            if stat == 'Mean':
                value = data.mean()
            elif stat == 'Median':
                value = data.median()
            elif stat == 'Mode':
                value = mode(data, keepdims=True).mode[0] if not data.empty else np.nan
            elif stat == 'Range':
                value = data.max() - data.min()
            elif stat == 'Stdev':
                value = data.std()
            elif stat == 'Variance':
                value = data.var()
            elif stat == 'IQR':
                value = data.quantile(0.75) - data.quantile(0.25)
            elif stat == '25%':
                value = data.quantile(0.25)
            elif stat == '50%':
                value = data.quantile(0.50)
            elif stat == '75%':
                value = data.quantile(0.75)
            elif stat == 'Min':
                value = data.min()
            elif stat == 'Max':
                value = data.max()
            elif stat == 'Skewness':
                value = skew(data)
            elif stat == 'Kurtosis':
                value = kurtosis(data)
            else:
                value = np.nan
            row.append(round(value, 2) if not pd.isnull(value) else np.nan)
        stats_table.append(row)

    stats_df = pd.DataFrame(stats_table, columns=numeric_cols, index=stat_names)
    return stats_df

# Uso:
# Asegúrate de tener tu DataFrame listo y sin valores nulos:
# df_vacios_eliminados = df.dropna()
stats_df = calculate_statistics(df_vacios_eliminados)

# Mostrar la tabla completa
print(stats_df)
```
